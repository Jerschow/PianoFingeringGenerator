```python
from music21 import *
```

```python
def give_fingerings(song):
    new_stream = song.parts.stream()
    elements = list(new_stream.elements)
    for part_staff_index in range(len(elements)):
        part_staff_is_tuple = False
        if isinstance(elements[part_staff_index], tuple):
            elements[part_staff_index] = list(elements[part_staff_index])
            part_staff_is_tuple = True
        part_staff = elements[part_staff_index]
        for measure_index in range(len(part_staff)):
            measure_is_tuple = False
            if isinstance(part_staff[measure_index], tuple):
                part_staff[measure_index] = list(part_staff[measure_index])
                measure_is_tuple = True
            measure = part_staff[measure_index]
            if isinstance(measure, stream.Measure):
                for nt_index in range(len(measure)):
                    nt = measure[nt_index]
                    if isinstance(nt, note.Note) or isinstance(nt, chord.Chord):
                        elements[part_staff_index][measure_index][nt_index].lyric = '1'
            if measure_is_tuple:
                part_staff[measure_index] = tuple(part_staff[measure_index])
        if part_staff_is_tuple:
            elements[part_staff_index] = tuple(elements[part_staff_index])
    new_stream.elements = tuple(elements)
    new_stream.show()
```

```python
song = converter.parse('samples/mary_had_a_little_lamb.mxl')
```

```python
give_fingerings(song)
```
